```python
import numpy as np
import json
import matplotlib.pyplot as plt
import pandas as pd
import os
import time
from tkinter import Tcl
from tqdm.notebook import tqdm
from scipy.interpolate import interp1d
```

```python
def readDict(path):
    if ".json" not in path:
        readName = path+'.json'
    else:
        readName = path
    try:
        with open(readName, "r") as infile:
            dictValues = json.load(infile)
            return(convert_lists_to_arrays(dictValues))
    except IOError as e:
        print(e)
        return('None')
    except ValueError as e:
        print(e)
        return('None')
    
def convert_lists_to_arrays(data):
    if isinstance(data, list):
        return np.array(data)
    if isinstance(data, dict):
        return {key: convert_lists_to_arrays(value) for key, value in data.items()}
    return data
```

```python
fig, ax = plt.subplots(2,2, figsize= (10*2,9*2))

path= "LCEHL_EOS_posterior_samples_PSR/"
alpha= 0.01

for file in tqdm(Tcl().call('lsort', '-dict', os.listdir(path))):
    if file.endswith(".json"): dat= readDict(path+file)
    else: continue

    M_array= dat["M"]
    R_array= dat["R"]
    Lamda_array= dat["Lambda"]
    f_array= dat["NRO_GR"]["f"]
    tao_array= dat["NRO_GR"]["DT"]

    ax[0,0].plot(R_array, M_array,"-", color="k",alpha=alpha)
    ax[0,1].plot(M_array, Lamda_array,"-", color="k",alpha=alpha)
    ax[1,0].plot(M_array, f_array,"-", color="k",alpha=alpha)
    ax[1,1].plot(M_array, tao_array,"-", color="k",alpha=alpha)

ax[0,0].set_xlabel(r"$R \ (\mathrm{km})$", labelpad= 10)
ax[0,0].set_ylabel(r'$M \ (M_\odot)$', labelpad= 10)
ax[0,0].grid(alpha = 0.4)

ax[0,1].set_xlabel(r'$M \ (M_\odot)$', labelpad= 8)
ax[0,1].set_ylabel(r"$\mathrm{\Lambda}$", labelpad= 10)
ax[0,1].set_yscale("log")
ax[0,1].grid(alpha = 0.4)

ax[1,0].set_xlabel(r'$M \ (M_\odot)$', labelpad= 8)
ax[1,0].set_ylabel(r"$f \ \mathrm{(kHz)}$", labelpad= 10)
ax[1,0].grid(alpha = 0.4)
ax[1,0].set_ylim(0.5,3)

ax[1,1].set_xlabel(r'$M \ (M_\odot)$', labelpad= 8)
ax[1,1].set_ylabel(r"$\tau \ \mathrm{(s)}$", labelpad= 5)
ax[1,1].set_yscale("log")
ax[1,1].grid(alpha = 0.4)
ax[1,1].set_ylim(0.1,3)

plt.show()
```
